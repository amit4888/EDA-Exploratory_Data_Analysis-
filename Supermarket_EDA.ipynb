```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt    # it is used for data visualization
import seaborn as sns
```

```python
df = pd.read_csv("supermarket.csv")
df
```

```python
# Shape of the dataset
df.shape
```

# Data Cleaning

```python
# Check null values

df.isnull().sum()
```

```python
# Check data types and missing values
df.info()
```

```python
# Note
# change column type --> Order Data and Ship Data 
# null values : postal code
```

```python
# Statistics summary
df.describe()
```

```python
## detect duplicate
df[df.duplicated()]

# So here, we can see there is no duplicate values in our dataset.
```

```python
# Change data type 
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
# Now, we can check dtype of Order Date and Ship Date has changed.
df.info()
```

```python
# This ensures that all postal codes are stored as integers instead of floating-point numbers.

df["Postal Code"] = df["Postal Code"].astype("Int64")
```

```python
df.info()
```

```python
df = df.dropna()  # Removes rows containing NaN values
```

```python
# I'm again checking. So here we can see there is no any missing values
df.isnull().sum()
```

# EDA (Exploratory Data Analysis)

```python
# Count unique values in Ship Mode
ship_mode_counts = df['Ship Mode'].value_counts()

plt.figure(figsize=(5,7))

ship_mode_counts.plot.pie(autopct='%1.1f%%',startangle=140)

# Add title
plt.title("Distribution of Ship Modes")

# Show plot
plt.show()
```

```python
# Count unique values in Ship Mode
ship_mode_counts = df['Segment'].value_counts()

plt.figure(figsize=(5,7))

ship_mode_counts.plot.pie(autopct='%1.1f%%',startangle=140)

# Add title
plt.title("Distribution of Segement")

# Show plot
plt.show()
```

```python
# Count unique values in Ship Mode
ship_mode_counts = df['Region'].value_counts()

plt.figure(figsize=(5,7))

ship_mode_counts.plot.pie(autopct='%1.1f%%',startangle=140)

# Add title
plt.title("Distribution of Regions")

# Show plot
plt.show()
```

```python
# Count unique values in Ship Mode
ship_mode_counts = df['Category'].value_counts()

plt.figure(figsize=(5,7))

ship_mode_counts.plot.pie(autopct='%1.1f%%',startangle=140)

# Add title
plt.title("Distribution of Category")

# Show plot
plt.show()
```

```python
# Extract year
df['Ship Date'] = df['Ship Date'].dt.year

df['Order Date'] = df['Order Date'].dt.year
```

```python
# Drop unnecessary columns

df.drop(columns = ["Order ID","Row ID","Customer ID","Product ID"],inplace = True)
```

```python
# Set figure size
plt.figure(figsize=(20,6))

# Create Countplot 
sns.countplot(x = df['State'], order=df['State'].value_counts().head(10).index, palette="Reds")

# Add title
plt.title("Count of Orders by State")

# Show plot
plt.show()
```

```python
# Note: .index
# Extracts only the state names.
# This gives an ordered list of states based on their count
```

```python
# Note : Order
# By default, sns.countplot() plots categories in random order.
```

```python
# Set figure size
plt.figure(figsize=(20,7))

# Create Countplot 
sns.countplot(x = df['City'], order=df['City'].value_counts().head(15).index, palette="Set2")

# Add title
plt.title("Count of Orders by City")

# Show plot
plt.show()
```

```python
# Set figure size
plt.figure(figsize=(20,6))

# Create Countplot 
sns.countplot(x = df['Sub-Category'], order=df['Sub-Category'].value_counts().index, palette="Blues")

# Add title
plt.title("Count of Orders by Sub-Category")

# Show plot
plt.show()
```
